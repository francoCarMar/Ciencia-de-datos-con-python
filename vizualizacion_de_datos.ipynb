```python
# Load data
import pandas as pd
nombres = ['preg', 'plas', 'pres','skin','test','mass','pedi','age','class']
data = pd.read_csv('data/pima-indians-diabetes.csv', names = nombres)
data
```


# Histogramas

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 10))
ax = fig.gca()
data.hist(ax = ax)
```

```python
%matplotlib inline
# Univariate Histograms
import matplotlib.pyplot as plt
import seaborn as sns
f, axes = plt.subplots(3, 3, figsize = (14,14))
sns.histplot(data['preg'], ax=axes[0,0])
sns.histplot(data['plas'], ax=axes[0,1])
sns.histplot(data['pres'], ax=axes[0,2])
sns.histplot(data['skin'], ax=axes[1,0])
sns.histplot(data['test'], ax=axes[1,1])
sns.histplot(data['mass'], ax=axes[1,2])
sns.histplot(data['pedi'], ax=axes[2,0])
sns.histplot(data['age'], ax=axes[2,1])
sns.histplot(data['class'], ax=axes[2,2])
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16, 16))
ax = fig.gca()
data.plot(ax = ax, kind = 'density', subplots = True, layout = (3, 3), sharex = False)
plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
f, axes = plt.subplots(3, 3, figsize = (14,14))
sns.kdeplot(data['preg'], ax=axes[0,0])
sns.kdeplot(data['plas'], ax=axes[0,1])
sns.kdeplot(data['pres'], ax=axes[0,2])
sns.kdeplot(data['skin'], ax=axes[1,0])
sns.kdeplot(data['test'], ax=axes[1,1])
sns.kdeplot(data['mass'], ax=axes[1,2])
sns.kdeplot(data['pedi'], ax=axes[2,0])
sns.kdeplot(data['age'], ax=axes[2,1])
sns.kdeplot(data['class'], ax=axes[2,2])
plt.show()
```

```python
# Box plots 
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16, 16))
ax = fig.gca()
data.plot(ax = ax, kind = 'box', subplots = True, layout = (3, 3), sharex = False, sharey = False)
plt.show()
```

```python
# Con seaborn
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
f, axes = plt.subplots(3, 3, figsize = (14,14))
sns.boxplot(y=data['preg'], ax=axes[0,0])
sns.boxplot(y=data['plas'], ax=axes[0,1])
sns.boxplot(y=data['pres'], ax=axes[0,2])
sns.boxplot(y=data['skin'], ax=axes[1,0])
sns.boxplot(y=data['test'], ax=axes[1,1])
sns.boxplot(y=data['mass'], ax=axes[1,2])
sns.boxplot(y=data['pedi'], ax=axes[2,0])
sns.boxplot(y=data['age'], ax=axes[2,1])
sns.boxplot(y=data['class'], ax=axes[2,2])
plt.show()
```

# Matriz de correlaciÃ³n

```python
%matplotlib inline
import numpy as np
correlations = data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_xticklabels(nombres)
ax.set_yticklabels(nombres)
```

```python
%matplotlib inline
correlation = data.corr()
plt.figure(figsize = (10, 10))
ax = sns.heatmap(correlation, vmax=1, square=True, annot=True, cmap='viridis')
plt.title('Matriz de correlaciÃ³n')
plt.show()
```
